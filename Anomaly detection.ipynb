# Anomaly detection

## Importing required libraries and Sklearn dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

## MakeMoon dataset from Sklearn for anomaly detection.

```python
X1,y1 = make_moons(n_samples=750,shuffle=True, noise = 0.1)
```

## Scatter plot using Matplotlib to identity the data points of X Since y is the output which need to be excluded or hidden

```python
plt.scatter(X1[:,0],X1[:,1])
```

## Data clustering algorithm proposed by Martin Ester, Hans-Peter Kriegel, Jörg Sander and Xiaowei Xu 

Why DBSCAN

Real life data, Clusters can be of arbitrary shape and Data may contain noise where (K-means, PAM clustering) and hierarchical clustering finds difficults to correctly classify the outliers.

```python
from sklearn.cluster import DBSCAN
```

## Create an instance of the DBSCAN clustering algorithm from the scikit-learn library, with the eps parameter.
**EPS = A crucial hyperparameter that determines the maximum distance between two points for them to be considered part of the same neighborhood or cluster.**

Predicted array consist of -1 values which are consdier as outliers.


```python
dbscan = DBSCAN(eps = 0.10)
dbscan.fit_predict(X1)
```

## Labale check of the data points

**Arrya of datapoints groups are obtained. were - 1 denotes Outliers**

```python
dbscan.labels_
```

## Scatter plot for Visualization of Datapoints and outliers

**Group 1 - Green**

**Group 2 - Blue**

**Group 3 - Yellow**

**Outliers - Black**

```python
plt.scatter(X1[:,0],X1[:,1],c=dbscan.labels_)
```
